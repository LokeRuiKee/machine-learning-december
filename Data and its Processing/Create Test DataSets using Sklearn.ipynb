This tutorial creates datasets using sklearn library. In this tutorial we're playing with 3 different methods in the 'sklearn.dataset' module: make_blobs, make_moon, make_circle. 

Learning material: https://www.geeksforgeeks.org/python-create-test-datasets-using-sklearn/

# sklearn.datasets.make_blobs

```python
# import library
from sklearn.datasets import make_blobs

# matplotlib for plotting
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
# just some styling (other matplotlib styling: default, ggplot, bmh ...)
style.use("fivethirtyeight")

x, y = make_blobs(n_samples = 100, centers = 3, cluster_std = 1, n_features = 2)

plt.scatter(x[:, 0], x[:, 1], s = 40, color = 'g')
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
plt.clf()
```

make_blobs: used to create a synthetic dataset.
other sklearn.dataset methods i.e. make_circle.
more details: https://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets

we're now creating a dataset of 100 samples, 3 centers, a cluster standard deviation of 1, and 2 features.
Low standard deviation will create a relatively tight cluster and the points will be closer to center. High standard deviation will create a more spread out cluster and the points are distributed over a larger area.

 x[:, 0] means selecting all rows (:) of the array x and only the elements in the first column (0). It's one of the feature. Another feature being  x[:, 1].
 s = 40 sets the size of the points.
 
 plt.clf clears the current figure, preparing the plot canvas for a new plot. This is useful if we want to create multiple plots in the same script.
 
 The <Figure size 640x480 with 0 Axes> is generated by 'plt.clf()'. It's saying that the figure has been cleared and there are currently no axes (plots) in the figure.


# sklearn.datasets.make_moon

```python
from sklearn.datasets import make_moons

x, y = make_moons(n_samples = 1000, noise = 0.1)
plt.scatter(x[:, 0], x[:, 1], s = 40, color = 'b')
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
plt.clf()
```

# sklearn.datasets.make_circle

```python
from sklearn.datasets import make_circles

# trying with other styles in matplotlib 
style.use("ggplot")

x, y = make_circles(n_samples = 100, noise = 0.02)
plt.scatter(x[:, 0], x[:,1 ], s = 40, color = 'r')
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
plt.clf()
```

# Multi-Class Classification

Example 1: the data generated are blobs to utilize for clustering

*blob = binary large object (aka long strings of data)

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
x, y = make_blobs(n_samples=500, centers=3, n_features=2, random_state=42)

plt.scatter(x[:,0],x[:, 1], c=y)
plt.show()
```

About random_state, you ensure that every time you run this code with the same seed, you get the same dataset. This is useful for reproducibility in research or testing.

About c=y, the c is colour. c=y means that the color of each point is determined by the corresponding values in the y array, which represents the cluster labels.

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

Example 3: A random multi-label classification data is created by the function make make_multilabel_classification() 

```python
from sklearn.datasets import make_multilabel_classification
import pandas as pd
import matplotlib.pyplot as plt

x, y =make_multilabel_classification(n_samples=500, n_features=2, n_classes=2, n_labels=2, allow_unlabeled=True, random_state=23)

# create pandas dataframe from generated dataset
df = pd.concat([pd.DataFrame(x,columns=['X1', 'X2']), pd.DataFrame(y, columns=['Label1', 'Label2'])], axis=1)

display(df.head())

plt.scatter(df['X1'], df['X2'], c=df['Label1'])
plt.show()
```

*A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

Example 2:  Multilabel feature using make_sparse_uncorrelated()

```python
from sklearn.datasets import make_sparse_uncorrelated
import matplotlib.pyplot as plt

# generate 1d Regression dataset
x, y = make_sparse_uncorrelated(n_samples=100, n_features=4, random_state=23)

plt.figure(figsize=(12,10))

for i in range(4):
    plt.subplot(2,2, i+1)
    plt.scatter(x[:, i], y)
    plt.xlabel('X'+str(i+1))
    plt.ylabel('Y')

plt.show()
```

The loop for i in range(4): iterates over each feature, creating a subplot for visualizing each feature against the target variable.

plt.subplot(2, 2, i+1): Divides the figure into a 2x2 grid and selects the i-th subplot for the current feature.

plt.scatter(x[:, i], y): Plots a scatter plot where the x-axis corresponds to the i-th feature (x[:, i]) and the y-axis corresponds to the target variable (y).

plt.xlabel('X' + str(i+1)): Sets the x-axis label for the subplot.

plt.ylabel('Y'): Sets the y-axis label for the subplot.

```python
# same thing as above but displaying the data with 3D scatter plot.

from sklearn.datasets import make_sparse_uncorrelated
import matplotlib.pyplot as plt

# generate 1d Regression dataset
x, y = make_sparse_uncorrelated(n_samples=100, n_features=4, random_state=23)

# Create a 3D scatter plot
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=y)

# Set labels for the axes
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

plt.show()
```

Q: the generated dataset has 4 features right? does that mean it's a 4D dataset?

A: In this code, n_features=4 specifies that each sample in the dataset has four features. This means that each data point is characterized by a vector of four values. In a geometric sense, you can think of this dataset as existing in a four-dimensional space.

The term "4D dataset" indicates that each data point has four dimensions, with each dimension corresponding to one of the four features. While it's challenging to visualize directly in a plot due to our limited ability to visualize beyond three dimensions, various techniques, such as dimensionality reduction or slicing, can be used to gain insights into the relationships between features in a high-dimensional dataset.

Q: but why in the line of x, y =, it's saying that to generate 1D regression dataset? isn't it a 4D dataset

A: 

The term "1D regression dataset" might be a bit confusing in this context, and it seems like there might be a misunderstanding in the naming of the function parameters. Let's clarify the terminology:

In this code, the function make_sparse_uncorrelated is used to generate a dataset for a 1D regression problem. The term "1D regression" typically refers to the fact that the target variable (y in this case) is a one-dimensional array. It doesn't directly imply the dimensionality of the input features.

However, the function is configured to generate a dataset with four features (n_features=4). So, technically, it's a 4D dataset in terms of the input features. The "1D" in "1D regression" refers to the fact that the target variable is a one-dimensional array (representing the regression output).

To avoid confusion, it might be more accurate to say that the code generates a dataset for a regression problem with four features rather than specifically calling it a "1D regression dataset." The target variable is one-dimensional, but the input features have four dimensions.

Example: 3  Multilabel feature using make_friedman2()

```python
from sklearn.datasets import make_friedman2
import matplotlib.pyplot as plt

x,y = make_friedman2(n_samples = 100, random_state=23)

plt.figure(figsize=(12,10))

for i in range(4):
    plt.subplot(2,2, i+1)
    plt.scatter(x[:, i], y)
    plt.xlabel('X'+str(i+1))
    plt.ylabel('Y')
    
plt.show()
```

x (input feature) and y (target variable that you want to predict)

The scatter plots aim to visualize the relationship between the 4 feature and the target variable for different perspectives.

```python
# display the generated dataset in a table and a single digram

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_sparse_uncorrelated

# Generate 1D Regression dataset
x, y = make_sparse_uncorrelated(n_samples=100, n_features=4, random_state=23)

# Create a DataFrame for the dataset
df = pd.DataFrame(x, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4'])
df['Target'] = y

# Display the dataset table
print("Generated Dataset:")
print(df.head())

# Create a single diagram (scatter plot) for visualization
plt.figure(figsize=(10, 6))

# Scatter plot of Feature1 against Target
plt.subplot(2, 2, 1)
plt.scatter(df['Feature1'], df['Target'])
plt.xlabel('Feature1')
plt.ylabel('Target')

# Scatter plot of Feature2 against Target
plt.subplot(2, 2, 2)
plt.scatter(df['Feature2'], df['Target'])
plt.xlabel('Feature2')
plt.ylabel('Target')

# Scatter plot of Feature3 against Target
plt.subplot(2, 2, 3)
plt.scatter(df['Feature3'], df['Target'])
plt.xlabel('Feature3')
plt.ylabel('Target')

# Scatter plot of Feature4 against Target
plt.subplot(2, 2, 4)
plt.scatter(df['Feature4'], df['Target'])
plt.xlabel('Feature4')
plt.ylabel('Target')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

In this example:

We use pandas to create a DataFrame (df) with columns for each feature and the target variable.

We display the first few rows of the dataset using df.head() to create a table.

We create a single diagram (subplot) with scatter plots of each feature against the target variable.
